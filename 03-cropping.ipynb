## Cropping



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
import os

# اتصال به Google Drive
drive.mount('/content/drive')

# مسیر فایل ورودی (لطفاً مسیر دقیق را جایگزین کنید)
image_path = '/content/drive/My Drive/images/input.jpg'

# خواندن تصویر
image = cv2.imread(image_path)

# بررسی اینکه تصویر به درستی خوانده شده باشد
if image is None:
    print("خطا در خواندن تصویر! مسیر فایل را بررسی کنید.")
else:
    # نمایش تصویر با matplotlib
    plt.imshow(image[..., ::-1])  # تبدیل BGR به RGB
    plt.axis('off')
    plt.show()
```

```python

height, width = image.shape[:2]

# Simply use indexing to crop out the rectangle we desire
cropped = image[250:400 , 250:550]

plt.imshow(cropped[...,::-1])
```
